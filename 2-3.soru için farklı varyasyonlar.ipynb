```python
#Kütüphanemiz import edelim
import pandas as pd
import numpy as np
```

```python
#Verimizi import edelim
data=pd.read_csv('car_price_prediction_.csv')
```

```python
#Veri setinin 5 satırını alalım
data.head()
```

```python
# İncelenecek sütunlar
columns_normalize = ['Engine Size', 'Price']

# Z-Score Normalizasyonu
z_score_normalize = (data[columns_normalize] - data[columns_normalize].mean()) / data[columns_normalize].std()

# Min-Max Normalizasyonu
min_max_normalize = (data[columns_normalize] - data[columns_normalize].min()) / (data[columns_normalize].max() - data[columns_normalize].min())
```

```python
print(f"Z-Score:{z_score_normalize}")
```

```python
print(f"MAX-MİN:{min_max_normalize}")
```

2-Detaylı Değişken Analizi
a)Üretim Yılı(Year)
*Ortalama(Mean):2011 araç modellerin ortalamasını göstermektedir.
*En Eski Araç:2000 yılına ait araç en eski aracımızdır.
*En Yeni Araç:2023 yılına ait araç en yei modelimizidir.
//Analiz:Araçların çoğu ağırlıkla 2005-2018 arasında üretilmiştir.Daha yeni araçların fiyat bakımından eski araçlara göre farklılıklar gösterdiği görülmektedir.


b)Motor Hacmi(Engine Size)
*Ortalama:3.47 litre araç motor haciminin ortalasıdır.
*Standart Sapma:1.43 litre 
*Aralık:1.0 Litre(en kücük) ve 6.0 litre(büyük motor)
Analiz:Motor hacmi genelde orta segmentte agırlık göstermiştir.(3.4 litre medyan değeri) Daha büyük motor hacimli araçlar genellikle lüks segmente aittir.

c. Kilometre (Mileage)
Ortalama (Mean): 149,749.84 km
Standart Sapma (Std): 87,919.95 km
Aralık: 15 km (neredeyse hiç kullanılmamış) ile 299,967 km (aşırı kullanılmış).
Analiz: Verideki kilometre dağılımı geniştir. Bu, araçların yaşları ve kullanım yoğunluğundaki çeşitliliği yansıtmaktadır. Yüksek kilometreye sahip araçlar genellikle daha düşük fiyatlandırılır.

d. Fiyat (Price)
Ortalama (Mean): 52,638.02 USD
Standart Sapma (Std): 27,295.83 USD
Aralık: 5011.27 USD (ekonomik segment) ile 99,982.59 USD (lüks segment).
Analiz: Fiyatlar arasında geniş bir varyasyon gözlemlenmiştir. Medyan fiyat (53,485.24 USD), veri setinin genelinde orta segment araçların yaygın olduğunu göstermektedir.

4. Öneriler
Fiyat ve Motor Hacmi Analizi:

Daha büyük motor hacmine sahip araçların daha yüksek fiyatlandırıldığı göz önüne alınarak, lüks segment araçlara yönelik fiyatlandırma stratejisi gözden geçirilebilir.
Kilometreye Göre Fiyatlandırma:

Kilometresi yüksek araçların fiyatının düşürüldüğü varsayımı doğrulanmalı ve bu ilişki üzerinden fiyat tahmin modeli geliştirilebilir.
Segmentasyon:

Araçları fiyat, motor hacmi ve kilometreye göre ekonomik, orta segment ve lüks olarak gruplandırmak, satış stratejilerinde hedef odaklı yaklaşım sunabilir.
Trend Analizi:

Üretim yılına göre araç fiyatlarının değişimini incelemek, piyasa trendlerini anlamada faydalı olacaktır.


```python
import seaborn as sns
import matplotlib.pyplot as plt

#Seaborn görselleştirmeleri
columns_to_plot=['Engine Size','Mileage','Price']

#Her sütun için dağılım grafiği

for column in columns_to_plot:
    plt.figure(figsize=(10,6))
    sns.histplot(data[column],kde=True,bins=30,color="blue")
    plt.title(f"{column} Disruibiton",fontsize=12)
    plt.xlabel(column)
    plt.ylabel('Frequency')
    
```
