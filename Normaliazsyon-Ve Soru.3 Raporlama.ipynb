```python
#Bu kısımda Büyük Veri Analitiğinde çıkacak olan 2.soru ve 3.soru için çözümleme ve raporlama yapacagız.Verilen verilere göre z-score 
#max-min normalizasyon yapacagız.3.soruda ise verilen verinin istatiksel raporlamasını oluşturacağız.
```

```python
#İlk başta kütüphaneyi import edelim.
import pandas as pd
```

```python
#Verimizi import edeceğiz.
data=pd.read_csv('ex1data2.txt')
```

```python
#İlk birkaç satırımızı görüntüleyelim
data.head()
```

```python
#Sütun isimlendirmeleri ile verimizi düzenleyelim
columns=['M^2','Room','Price']
data.columns=columns
```

```python
#Z-score hesaplamasını yapalım
z_score=(data-data.mean())/data.std()
print(f'Z-Score Değeri:{z_score}')
```

```python
#Max-Min Normalizasyonu yapalım
max_min=(data-data.min())/(data.max()-data.min())
print(f'Max-Min Score Hesaplama:{max_min}')
```
